붓꽃 데이터 시각화
막대 그래프 그리기
1. 꽃받침 길이의 시각화
1. 붓꽃 종류별 꽃받침 길이의 평균에 대한 시각화


3. 산점도 그래프 그리기
• 꽃받침의 길이와 꽃잎의 길이의 산점도를 시각화

```python
import matplotlib.pyplot as plt
plt.bar(iris_data.index[:10],iris_data.sepal_length[:10])  #데이터를 잘라서 보여줌. 너무 많아서
plt.title("Sepal Length Visualization")
plt.xlabel('Data')
plt.ylabel('sepal length')
plt.show()
```

```python
sns.barplot(x="species",y="sepal_length",hue="species",data=iris_data) # 붓꽃 종류별 꽃받침길이의 평균.씨본사용
plt.show()
```

```python
import matplotlib.pyplot as plt  # 붓꽃 종류별 꽃받침길이의 평균.matplotlib 사용
df3=iris_data.groupby('species').mean()
plt.bar(df3.index,df3['sepal_length'])
plt.title('species sepal length graph')
plt.xlabel('species')
plt.ylabel('petal length')
```
